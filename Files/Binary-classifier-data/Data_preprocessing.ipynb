```python
import numpy as np
import pandas as pd
```

```python
file = 'train.csv'
proc_file = 'processed_train.csv'
```

# Start preprocessing dataset

```python
df = pd.read_csv(file, names=[f'col_{i}' for i in range(100)])
```

```python
df
```

```python
df.describe()
```

```python
[df.iloc[:,i].sum() for i in range(df.shape[1])]
```

# Filling in NaN values with means

```python
for i in range(df.shape[1]):
    df.iloc[:, i].fillna(df.iloc[:, i].mean(), inplace=True)
```

```python
[df.iloc[:,i].sum() for i in range(df.shape[1])]
```

```python
df.to_csv("processed_train.csv", index=False)
```
